# Exploratory Data Analysis

EDA is a crucial step in data analysis that involves studying, exploring, and visualizing information to derive important insights. To find *patterns*, *trends*, and *relationships* in the data, it makes use of statistical tools and visualizations. This helps to formulate hypotheses and direct additional investigations. This procedure improves data comprehension and provides information for further modeling decisions.
EDA is used for gaining a better understanding of data aspects like: 
- main **features** of data 
- **variables and relationships** that hold between them 
- Identifying **which variables are important** for our problem 

**Exploratory data analysis methods include:** 

1. Reading dataset
2. Analyzing the data
3. Checking for the duplicates
4. Missing Values Calculation
5. Exploratory Data Analysis
-   Univariate Analysis
-   Bivariate Analysis
-   Multivariate Analysis


To obtain a comprehensive understanding of the data, analysts use a variety of EDA approaches, including summary statistics, correlation analysis, and data visualization using tools like box plots, scatter plots, and histograms

### Univariate Analysis

In Univariate analysis, plotting the right charts can help us better understand the data, which is why data visualization is so important. Matplotlib and Seaborn libraries are used in this post to visualize our data.

#### **Numerical Features**


Distribution Shape: Histograms help visualize the distribution of numerical data, showing how data points are spread across different ranges. You can observe whether the data is skewed (left or right), symmetrical, or has any unusual peaks (modes).

Central Tendency and Spread: By visualizing a histogram, you can get a sense of the central tendency (mean or median) and the spread (variance or standard deviation) of the data.

Outliers: Histograms can help identify outliers, which are values that deviate significantly from other observations in the data.

Frequency of Values: They show how frequently each range of values occurs, which helps in understanding the density of data points in specific intervals

```python
quality_counts = wine_df['quality'].value_counts()

# Using Matplotlib to create a count plot
plt.figure(figsize=(8, 6))
plt.bar(quality_counts.index, quality_counts, color='pink')
plt.title('Count Plot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

##### Kernel density

Kernel density plots are a way to estimate and visualize the probability density function of a continuous random variable. Unlike histograms, which use bins to estimate density, kernel density plots smooth the data using a kernel function (like a Gaussian) to create a continuous curve

```python
# Set Seaborn style
sns.set_style("darkgrid")

# Identify numerical columns
numerical_columns = wine_df.select_dtypes(include=["int64", "float64"]).columns

# Plot distribution of each numerical feature
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(wine_df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(wine_df[feature].skew(), 2)}")

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

#### **Categorical Features**


Here's why visualizing categorical features is useful:

Frequency Distribution: Bar plots show the frequency or count of each category, helping you see which categories are more or less common.

Comparisons: You can easily compare the size of different categories to identify the most or least common ones.

Patterns and Trends: They can reveal patterns or trends within categorical data, such as seasonal trends or popular categories.

##### Swarm plot

```python
# Assuming 'df' is your DataFrame
plt.figure(figsize=(10, 8))

# Using Seaborn to create a swarm plot
sns.swarmplot(x="quality", y="alcohol", data=wine_df, palette='viridis')

plt.title('Swarm Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

This plot depicts that the higher point density in specific regions shows the concentration indicating where the majority of data points cluster. The points isolated and are far away from the clusters shows the outliers.

### Bivariate analysis

When doing a bivariate analysis, two variables are examined simultaneously in order to look for patterns, dependencies, or interactions between them. Understanding how changes in one variable may correspond to changes in another requires the use of this statistical method.

Bivariate analysis allows for a thorough comprehension of the interdependence between two variables within a dataset by revealing information on the type and intensity of associations.

##### Pair plot

```python
# Set the color palette
sns.set_palette("Pastel1")

# Assuming 'df' is your DataFrame
plt.figure(figsize=(10, 6))

# Using Seaborn to create a pair plot with the specified color palette
sns.pairplot(wine_df)

plt.suptitle('Pair Plot for DataFrame')
plt.show()
```

- Distribution: Diagonal plots tell you how each variable is distributed.
- Relationships: Scatter plots reveal how variables are related.
- Symmetry: Mirror images indicate consistent relationship

##### Violin plot

```python
# Assuming 'df' is your DataFrame
wine_df['quality'] = wine_df['quality'].astype(str)  # Convert 'quality' to categorical

plt.figure(figsize=(10, 8))

# Using Seaborn to create a violin plot
sns.violinplot(x="quality", y="alcohol", data=wine_df, palette={
               '3': 'lightcoral', '4': 'lightblue', '5': 'lightgreen', '6': 'gold', '7': 'lightskyblue', '8': 'lightpink'}, alpha=0.7)

plt.title('Violin Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

- Width of Violin: Indicates density of data.
- Shape: Reveals distribution and modes. Symmetrical plot indicates a balanced distribution
- Peak or bulge in the violin plot represents most common value in distribution
- Longer tails indicate great variability
- Median Line: Shows the central tendency.
- Comparison: Helps understand how distributions vary across categories

##### Box plot

```python
#plotting box plot between alcohol and quality
sns.boxplot(x='quality', y='alcohol', data=wine_df)
```

- **Box Plot:** Visualizes the distribution, central tendency, and variability of data.
- **Box (IQR):** Shows the range where the middle 50% of data lies.
- **Median:** Indicates the central value of the data.
- **Whiskers:** Extend to show the range of the data within 1.5 * IQR.
- **Outliers:** Identifies extreme values outside the whiskers.

### Multivariate analysis

Multivariate analysis explores how three or more variables interact in a dataset. It reveals patterns and relationships between these variables using methods like factor analysis, principal component analysis, and multivariate regression. This approach helps uncover how variables are related and supports decision-making in fields such as biology, economics, and marketing.

```python
# Assuming 'df' is your DataFrame
plt.figure(figsize=(15, 10))

# Using Seaborn to create a heatmap
sns.heatmap(wine_df.corr(), annot=True, fmt='.2f', cmap='Pastel2', linewidths=2)

plt.title('Correlation Heatmap')
plt.show()
```

- Values close to +1 indicate a strong positive correlation, -1 indicates a strong negative correlation and 0  suggests no linear correlation.
- Darker colors signify strong correlation, while light colors represents weaker correlations.
- Positive correlation variable move in same directions. As one increases, the other also increases.
- Negative correlation variable move in opposite directions. An increase in one variable is associated with a decrease in the other
