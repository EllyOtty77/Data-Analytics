```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available  # list of styling options
```

```python
plt.style.use('classic') # select chart style
```

## Line chart

```python
games = [1, 2, 3, 4, 5, 6]
goals = [3, 4, 5, 6, 7, 8]  
cards = [2, 2, 3, 3, 4, 1]

plt.plot(goals)
plt.xlabel('Games Played')
plt.ylabel('Goals Scored')
plt.title('Goals Scored per Game')
plt.show()
```

### Chart legend

```python
plt.plot(games, cards, label='Yellow Cards')
plt.xlabel('Games Played')
plt.ylabel('Cards accumulated')
plt.title('Cards per game')
plt.grid()
plt.legend()
plt.show()
```

```python
years = ['2020', '2021', '2022', '2023', '2024']
prices1 = [12, 15, 17, 20, 22]
prices2 = [10, 12, 16, 14, 17]

plt.plot(years, prices1, label='Glen prices')
plt.plot(years, prices2, label='Jamo prices')

plt.xlabel('Years')
plt.ylabel('Prices per bag')
plt.title('Prices over the years')
plt.legend(loc=2) # use loc to change legend location
plt.grid()
plt.show()
```

### Formatting and saving

```python
plt.plot(years, prices1, color="orange", marker='o', linestyle='--', label='Glen Prices')
plt.plot(years, prices2, color="green", marker='s', linestyle='-.', label='Jamo Prices')

plt.xlabel('Years')
plt.ylabel('Prices per bag')
plt.title('Prices over the years')
plt.legend(loc=2)  
plt.grid()

plt.savefig('line_chart.png') # Saving the chart
plt.show() 
```
