## Figures and subplots

Plots in matplotlib reside within a Figure object. You can create a new figure with plt.figure:

```python
fig = plt.figure()
fig
```

```python
fig = plt.figure() 
ax1 = fig.add_subplot(2, 2, 1) 
ax2 = fig.add_subplot(2, 2, 2) 
ax3 = fig.add_subplot(2, 2, 3)
```
