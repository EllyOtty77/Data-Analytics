# Exploratory Data Analysis

EDA is a crucial step in data analysis that involves studying, exploring, and visualizing information to derive important insights. To find *patterns*, *trends*, and *relationships* in the data, it makes use of statistical tools and visualizations. This helps to formulate hypotheses and direct additional investigations. This procedure improves data comprehension and provides information for further modeling decisions.
EDA is used for gaining a better understanding of data aspects like: 
- main **features** of data 
- **variables and relationships** that hold between them 
- Identifying **which variables are important** for our problem 

**Exploratory data analysis methods include:** 

1. Reading dataset
2. Analyzing the data
3. Checking for the duplicates
4. Missing Values Calculation
5. Exploratory Data Analysis
-   Univariate Analysis
-   Bivariate Analysis
-   Multivariate Analysis


To obtain a comprehensive understanding of the data, analysts use a variety of EDA approaches, including summary statistics, correlation analysis, and data visualization using tools like box plots, scatter plots, and histograms

### Univariate Analysis

In Univariate analysis, plotting the right charts can help us better understand the data, which is why data visualization is so important. Matplotlib and Seaborn libraries are used in this post to visualize our data.

#### **Numerical Features**


```python
# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# Create histograms for numerical features
numerical_features = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
housing_data[numerical_features].hist(figsize=(10, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features')
plt.show()
```


The histograms for numerical features (price, area, bedrooms, bathrooms, stories, parking) show the distribution of each feature.

*Price*: The distribution is right-skewed, indicating a few houses have very high prices.

*Area*: Also right-skewed, with most houses having a smaller area.

*Bedrooms*: Most houses have between 2 to 3 bedrooms.

*Bathrooms*: The majority have 1 to 2 bathrooms.

*Stories*: Most houses have 1 to 2 stories.

*Parking*: A large number of houses have no parking or one parking space.


#### **Categorical Features**


```python
# Create bar plots for categorical features
categorical_features = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=housing_data, x=feature, palette='viridis')
    plt.title(f'Count Plot of {feature}')
plt.tight_layout()
plt.show()
```

The bar plots for categorical features (mainroad, guestroom, basement, hotwaterheating, airconditioning, prefarea, furnishingstatus) indicate the frequency of each category.

*Mainroad*: Most houses are located on the main road.

*Guestroom*: Fewer houses have a guest room.

*Basement*: Around half of the houses have a basement.

*Hotwater Heating*: Very few houses have hot water heating.

*Airconditioning*: More houses have no air conditioning

*Preferred Area*: fewer houses are located in a preferred area.

*Furnishing Status*: Semi-furnished houses are the most common
